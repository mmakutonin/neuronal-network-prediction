```python
import numpy as np
import matplotlib.pyplot as plt
from util_functions import load_file, starting_run
```

### Notebook Parameters

```python
model_df = load_file('data_meta_pickled/pca_models_pickled_df.pickle')
```

### Notebook Functions

```python
def plot_model_explained_variance(model_df):
    for model_row in model_df.itertuples():
        model = load_file(model_row.model_path)
        plt.bar(range(model.n_components_), model.explained_variance_)
        plt.title(f"{model_row.subject} Stimulus {model_row.stimulus or 'All Stimuli'}: Total Explained Variance = {model_row.sum_explained_variance}")
        plt.show()
```

### Main Code Execution

```python
sub_model_df = model_df[model_df.stimulus.isna()]
plt.bar(sub_model_df.subject.apply(lambda x: x[-2:]), sub_model_df.sum_explained_variance)
plt.xlabel("Subject Number")
plt.ylabel("Proportion of Variance Explained")
plt.title("Proportion of explained variance per model")
plt.show()
```

```python
plt.bar(sub_model_df.subject.apply(lambda x: x[-2:]), sub_model_df["model_path"].apply( \
    lambda path: load_file(path).explained_variance_.sum()))
plt.xlabel("Subject Number")
plt.ylabel("Sum of Variance Explained")
plt.title("Sum of explained variance per model")
plt.show()
```

```python
plot_model_explained_variance(sub_model_df)
```

```python
plot_model_explained_variance(model_df[model_df["subject"] == "subject_r1"])
```
