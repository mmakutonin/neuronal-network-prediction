# Stimulus Continuity Analysis
This workflow showed that each stimulus was not applied continuously over the stimulation time period, and motivated Workflow 02-2.

```python
import numpy as np
import pandas as pd
from util_functions import load_file
```

```python
stimulus_list_df = load_file('data_meta_pickled/stimulus_df.pickle')
```

```python
for row in stimulus_list_df.itertuples():
    stimulus_timepoint_list = row.timepoints
    max_time = stimulus_timepoint_list.max()
    min_time = stimulus_timepoint_list.min()
    if row.timepoints_count != (max_time - min_time): # if the stimulus is not continuous, print.
        print(row.subject, ' ', row.stimulus)
        print(row.timepoints_count, ' ', (max_time - min_time))
```
